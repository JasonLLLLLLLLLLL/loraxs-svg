```python
import json
import random

# 定义颜色词汇列表
color_words = ['red','grey' ,'blue', 'green', 'yellow', 'black', 'white', 'gray', 'pink', 'orange', 'purple']
# 定义替换词列表，可以根据需要扩展
replacement_words = ['red','grey' ,'blue', 'green', 'yellow', 'black', 'gray', 'pink', 'orange', 'purple']

def replace_colors(caption):
    # 将caption中的颜色词替换成随机单词
    for color in color_words:
        if color in caption:
            # 随机选择一个替换词
            replacement_word = random.choice(replacement_words)
            # 替换所有出现的颜色词
            caption = caption.replace(color, replacement_word)
    return caption

def process_json_file(input_file_path, output_file_path):
    with open(input_file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)  # 假设输入是标准的JSON数组
    
    with open(output_file_path, 'w', encoding='utf-8') as outfile:
        for item in data:
            # 检查并替换caption中的颜色词
            if 'caption' in item:
                item['caption'] = replace_colors(item['caption'])
            # 将处理过的item写入.jsonl文件
            json.dump(item, outfile, ensure_ascii=False)
            outfile.write('\n')

# 使用函数处理文件
input_file_path = '/home/liuzhe/new-files/LoRA-XS/utils/dataset-1024-everypath-10-26.json'
output_file_path = '/home/liuzhe/new-files/LoRA-XS/utils/dataset-1024-everypath-10-26-color.jsonl'
process_json_file(input_file_path, output_file_path)
```

```python
import json
import random

def read_jsonl_file(file_path):
    """读取JSONL文件并返回一个包含所有记录的列表"""
    records = []
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            record = json.loads(line)
            records.append(record)
    return records

def write_jsonl_file(file_path, records):
    """将记录列表写入JSONL文件"""
    with open(file_path, 'w', encoding='utf-8') as file:
        for record in records:
            file.write(json.dumps(record, ensure_ascii=False) + '\n')

def main():
    # 文件路径
    input_file1_path = '/home/liuzhe/new-files/LoRA-XS/utils/dataset-1024-everypath-10-26-color.jsonl'
    input_file2_path = '/home/liuzhe/new-files/LoRA-XS/utils/dataset-1024-everypath-9-7.jsonl'
    output_file_path = 'path/to/your/output.jsonl'

    # 读取两个文件
    records1 = read_jsonl_file(input_file1_path)
    records2 = read_jsonl_file(input_file2_path)

    # 随机选取数据
    selected_records1 = random.sample(records1, 2331)
    selected_records2 = random.sample(records2, 5000)

    # 合并数据
    merged_records = selected_records1 + selected_records2

    # 写入新文件
    write_jsonl_file(output_file_path, merged_records)

if __name__ == '__main__':
    main()
```
